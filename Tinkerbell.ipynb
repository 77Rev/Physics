```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots

plt.style.use(['science','notebook','grid'])
```

```python
def tinker(a,b,c,d,x,y,r):
    xvalue = []
    yvalue = []
    
    def t(x,y):
          return x**2 - y**2 + a*x + b*y, 2*x*y+c*x+d*y
        
    for i in range(r):
        x,y = t(x,y)
        xvalue.append(x)
        yvalue.append(y)
        
    plt.figure(figsize=(15,10))
    plt.plot(xvalue,yvalue,'o',c='crimson',markersize=1.5)
    plt.ylabel('y',fontsize=16)
    plt.xlabel('x',fontsize=16)
    plt.title('Tinkerbell',fontsize=16)
    plt.grid(which='both')
    plt.show()
```

```python
tinker(0.87,-0.6,2,0.5,-0.72,-0.9,10000)
```
